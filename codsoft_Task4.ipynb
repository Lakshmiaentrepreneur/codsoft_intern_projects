```python
import tkinter as tk
from tkinter import ttk
import random
import string

def generate_password(length):
    # Define character sets for password complexity
    lowercase_letters = string.ascii_lowercase
    uppercase_letters = string.ascii_uppercase
    digits = string.digits
    special_characters = string.punctuation
    all_characters = lowercase_letters + uppercase_letters + digits + special_characters

    # Ensure length is at least 8 to include a variety of characters
    if length < 8:
        return "Password length should be at least 8 characters for better security."

    # Generate password
    password = ''.join(random.choice(all_characters) for _ in range(length))
    return password

def generate_and_display_password():
    # Get password length from the entry widget
    length = length_var.get()

    # Generate and display the password
    password = generate_password(length)
    if isinstance(password, str):
        result_label.config(text=password)
    else:
        result_label.config(text="Generated Password: " + password)

# Create the main window
root = tk.Tk()
root.title("Password Generator codsoft")

# Add a Frame to simulate the background color
main_frame = tk.Frame(root, background="#34495E", padx=10, pady=10)
main_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

length_label = ttk.Label(main_frame, text="Enter Password Length:")
length_label.grid(row=0, column=0, padx=5, pady=5, sticky=tk.W)

length_var = tk.IntVar(value=12)
length_entry = ttk.Entry(main_frame, textvariable=length_var, font=('Arial', 10))
length_entry.grid(row=0, column=1, padx=5, pady=5, sticky=tk.W)

generate_button = ttk.Button(main_frame, text="Generate Password", command=generate_and_display_password)
generate_button.grid(row=1, column=0, columnspan=2, pady=10, sticky=tk.W)

result_label = ttk.Label(main_frame, text="")
result_label.grid(row=2, column=0, columnspan=2, pady=5, sticky=tk.W)

# Configure the style for the button
style = ttk.Style()
style.configure("TButton", background="#3498DB", foreground="black")  # Black text on blue background
generate_button.configure(style="TButton")

# Run the main loop
root.mainloop()
```
